```python
#Q1
def count_pairs(lst, target_sum):
    count = 0
    length = len(lst)
    for i in range(length):
        for j in range(i + 1, length):
            if lst[i] + lst[j] == target:
                count += 1
    return count
lst = [2, 7, 4, 1, 3, 6]
target= 10
result = count_pairs(lst, target)
print(f"Number of pairs with sum {target}: {result}")


#Q2
def min_range(lst):
    min_val=lst[0]
    for i in lst:
        if i<min_val:
            min_val=i
    return min_val

def max_range(lst):
    max_val=lst[0]
    for i in lst:
        if i>max_val:
            max_val=i
    return max_val
   
lst=[5, 3, 8, 1, 0, 4]
if len(lst)<3:
    print("range determination not possible")
else:
    mini=min_range(lst)
    maxi=max_range(lst)
    res=maxi-mini
    print(f"range={res}")

#Q3    
def matrix_multiply(A, B):
    n = len(A) 
    result = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(n):
        for j in range(n):
            result[i][j] = sum(A[i][k] * B[k][j] for k in range(n))
    return result
def matrix_power(A, m):
    n = len(A)
    result = [[1 if i == j else 0 for j in range(n)] for i in range(n)]
    for _ in range(m):
        result = matrix_multiply(result, A)  
    return result
A = [
    [6, 4, 8],
    [9, 2, 3],
    [1, 7, 5]
]
m = 2
result = matrix_power(A, m)
print("Matrix A raised to the power m is:")
for row in result:
    print(row)
    
#Q4
def is_alphabetic(char):
    return (ord('a') <= ord(char) <= ord('z')) or (ord('A') <= ord(char) <= ord('Z'))
def highest_frequency(s):
    counts = [0] * 26
    for char in s:
        if is_alphabetic(char):
            char = char.lower()
            index = ord(char) - ord('a')
            counts[index] += 1
    max_count = 0
    max_char = '' 
    for i in range(26):
        if counts[i] > max_count:
            max_count = counts[i]
            max_char = chr(ord('a') + i)  
    return max_char, max_count
input_string = "hippopotamus"
char, count = highest_frequency(input_string)
print(f"The highest frequency is '{char}' with occurrence count {count}.")
```
